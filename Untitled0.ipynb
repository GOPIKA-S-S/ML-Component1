BINNING

```python
score_bins=[0,5,7,float('inf')]
bin_labels=['Low happiness','Medium Happiness','High happiness']
df['HappinessLevel']=pd.cut(df['Happiness Score'],bins=score_bins,labels=bin_labels,right=False)
print(df[['Happiness Score','HappinessLevel']])

```

```python
happiness_group_counts = df['HappinessLevel'].value_counts()
plt.bar(happiness_group_counts.index, happiness_group_counts.values)
plt.xlabel("Happiness Group")
plt.ylabel("Counts")
plt.title("Distribution of Happiness Groups")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

ONE HOT ENCODING

```python
#converting categorical to numerical values
df_encoded= pd.get_dummies(df, columns=["Region"])
print(df_encoded)
```

NORMALIZATION

```python
from sklearn.preprocessing import MinMaxScaler
required = ['Economy (GDP per Capita)']
minmax_scale= MinMaxScaler()
df[required] = minmax_scale.fit_transform(df[required])

```

CHI2 TEST

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['Region'], df['Happiness Rank'])

# Perform the Chi-Square test
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square Statistic:", chi2_stat)
print("P-value:", p_val)
print("Expected-value:",expected)
```

```python

contingency_table.plot(kind="bar",stacked=True)
plt.title("Region by Happiness Score")
plt.xlabel("Region")
plt.ylabel("Happiness Score")
plt.xticks(rotation=45)
plt.show()
```

COVARIANCE MATRIX

```python
economy= df['Economy (GDP per Capita)']
category_columns = df[['Health (Life Expectancy)']]

covariance_matrix = np.cov(economy, category_columns, rowvar=False)

print("Covariance Matrix:")
print(covariance_matrix)
```

HEAT MAP OF COVARIANCE MATRIX

```python
feature_names = ['economy', 'Life Expectancy']

plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(covariance_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
heatmap.set_xticklabels(feature_names, rotation=0)
heatmap.set_yticklabels(feature_names, rotation=0)

plt.title("Covariance Matrix Heatmap")
plt.show()
```
